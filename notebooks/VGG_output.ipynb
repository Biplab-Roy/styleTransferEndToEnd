```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
folders = [str(i) for i in range(1, 37)]
names = [i * 100 for i in range(1, 11)]
```

```python
for folder in folders:
    fig, axs = plt.subplots(2, 5, figsize=(20, 8))
    fig.suptitle(f"{folder}/style")
    for i in range(5):
        axs[0][i].set_title(f"Epoch: {names[2 * i]}")
        axs[0][i].imshow(Image.open(f"../results/{folders[0]}/style/{names[2 * i]}.png"))
        axs[1][i].set_title(f"Epoch: {names[2 * i + 1]}")
        axs[1][i].imshow(Image.open(f"../results/{folders[0]}/style/{names[2 * i + 1]}.png"))
        
    fig, axs = plt.subplots(2, 5, figsize=(20, 8))
    fig.suptitle(f"{folder}/content")
    for i in range(5):
        axs[0][i].set_title(f"Epoch: {names[2 * i]}")
        axs[0][i].imshow(Image.open(f"../results/{folders[0]}/content/{names[2 * i]}.png"))
        axs[1][i].set_title(f"Epoch: {names[2 * i + 1]}")
        axs[1][i].imshow(Image.open(f"../results/{folders[0]}/content/{names[2 * i + 1]}.png"))
plt.show()
```
